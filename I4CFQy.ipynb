```python
import numpy as np
from scipy.integrate import simpson
def calculate_relative_quantum_yield(I_sample, A_sample, n_sample, I_std, A_std, n_std, Q_ref):
    sample_wavelengths = np.linspace(460, 700, 100)
    sample_intensities = np.random.rand(100)*47
    std_wavelengths = np.linspace(460, 700, 100)
    std_intensities = np.random.rand(100)*444
    I_sample_integrated = simpson(sample_intensities, sample_wavelengths)
    I_std_integrated = simpson(std_intensities, std_wavelengths)
    Q_sample = Q_ref * (I_sample_integrated / I_std_integrated) * (A_std / A_sample) * ((n_sample**2) / (n_std**2))
    return round (Q_sample, 3)
```

```python
result = calculate_relative_quantum_yield (4704.07, 0.069, 1.36, 44470.7, 0.044, 1.36, 0.79)
```
